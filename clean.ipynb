```python
import pandas as pd
```

```python
df = pd.read_csv('scrap_data.csv')
df.head()
```

```python
df['deskripsi'] = df['deskripsi'].replace('\n', ' ', regex=True)
df['rating'] = df['rating'].replace('Belum ada rating', 0)
df['terjual'] = df['terjual'].str.replace(',', '.').str.replace('RB', '00')
df['terjual'] = df['terjual'].str.replace('.','').str.replace('+','0')
df['harga'] = df['harga'].str.extract(r'(Rp[\d\.]+)') # tambah setelah +, [^\d]*$ Jika mau ambil sebelah kanan
df['harga'] = df['harga'].str.replace('Rp', '').str.replace('.','')
```
