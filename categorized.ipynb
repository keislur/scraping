```python
# Import Library
import json
import pandas as pd
```

```python
df = pd.read_csv('clean_data.csv')

# Rename Colom sesuai dengan database
id_series = pd.Series(range(1, len(df)+1))
df = df.assign(id_kategori=1, stok = 500, id=id_series)

# Urutkan
order = ['id','id_kategori','kode','nama','harga','stok','penjual','lokasi','deskripsi','terjual','rating','rating_5','rating_4','rating_3','rating_2','rating_1']
df = df[order]
df.head()
```

```python
# convert the DataFrame to a JSON string
json_str = df.to_json(orient='records')

# Ubah jadi Objek Python
data = json.loads(json_str)
```

```python
# Kalkulasi rata-rata harga
prices = [float(d['harga']) for d in data]
avgPrice = sum(prices) / len(prices)

# Kategori berdasarkan price, rating, dan terjual
for d in data:
    price = float(d['harga'])
    rating = float(d['rating'])
    sales = int(d['terjual'])
    
    if price > avgPrice and rating > 4 and sales > 10 or price < avgPrice and rating > 4 and sales > 10:
        d['rekomendasi'] = 'Layak dibeli'
    elif price < avgPrice and rating > 4 and sales < 10 or price > avgPrice and rating < 4 and sales > 10:
        d['rekomendasi'] = 'Perlu dipertimbangkan'
    elif price > avgPrice and rating < 4 and sales < 10 or price < avgPrice and rating < 4 and sales < 10:
        d['rekomendasi'] = 'Harus dipertimbangkan'
    else:
        d['rekomendasi'] = 'Tidak layak dibeli'

# Save categorized data to categorized_data.json
with open('categorized_data.json', 'w') as f:
    json.dump(data, f)

df1 = pd.DataFrame(data)
df1.to_csv('categorized_data.csv', index = False)
```
